```python
# Step 0
!pip install english_words
from english_words import get_english_words_set
words = get_english_words_set(['web2'], lower=True)
```

```python
# Step 1 - Converting words into a list
wordlist = list(words)
# printing the amount of words into wordlist
print(len(wordlist))
```

```python
# Step 2 - Printing out the first ten words into wordlist
print(wordlist[:10])
```

```python
# Step 3 - Making a list in alphabetical order
alpha_list = sorted(wordlist)
print(alpha_list[:10])
```

```python
# Step 4 - words starting with a from the above list
# wanted to define the 10 word list I was working with instead of the entire word list
wordlist_10 = wordlist[:10]

count_a = len([word for word in wordlist if word.startswith('a')])
print("Number of words starting with 'a':", count_a)
```

```python
# Step 5 - words with q or without e
count_q = len([word for word in wordlist if 'q' in word])
count_without_e = len([word for word in wordlist if 'e' not in word])
fraction_without_e = count_without_e / len(wordlist)

print("Number of words with the letter 'q':", count_q)
print("Fraction of words without the letter 'e':", fraction_without_e)
```

```python
# Step 6 - importing numpy, converting lists to numpy arrays
import numpy as np

wordlist_np = np.array(wordlist)
alpha_list_np = np.array(alpha_list)

alphabet = [chr(i) for i in range(ord('a'),ord('z')+1)]

print(alphabet)
```

```python
# Step 7 - making bar graph based on starting letters
import matplotlib.pyplot as plt

word_count_per_letter = {letter: len([word for word in wordlist if word.startswith(letter)]) for letter in alphabet}

plt.figure(figsize=(9, 6))
plt.bar(word_count_per_letter.keys(), word_count_per_letter.values())
plt.xlabel('Letters')
plt.ylabel('Number of words')
plt.title('Number of Words from the 10-word List Starting with Each Letter')
plt.show()
```

```python
# Step 8 - six longest words in word list
words_sorted = sorted(wordlist, key=len, reverse=True)

six_longest_words = words_sorted[:6]

print("The six longest words in wordlist are:", six_longest_words)
```

```python
# Step 9 - histogram of word lengths in wordlist
import matplotlib.pyplot as plt

word_length_count = {length: len([word for word in wordlist if len(word) == length]) for length in range(1, max(map(len, wordlist)) + 1)}

plt.figure(figsize=(9, 6))
plt.bar(word_length_count.keys(), word_length_count.values())
plt.xlabel('Word Length')
plt.ylabel('Number of Words')
plt.title('A Histogram of Word Lengths in Wordlist')
plt.show()

most_common_length = max(word_length_count, key=word_length_count.get)
print("The most common length of a word in the wordlist is:", most_common_length)
```
